```python
import csv
file=open("Movie.csv","w")
writer=csv.writer(file)
writer.writerow(["Movie name","year","rating"])
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating"
response=requests.get(url)
soup=BeautifulSoup(response.content,'html.parser')
```

```python
movie_name=[]
year=[]
time=[]
rating=[]
meta_score=[]
votes=[]
gross=[]
```

```python
movie_data=soup.findAll('div',attrs={"class":"lister-item mode-advanced"})
```

```python
for store in movie_data:
    name=store.h3.a.text
    movie_name.append(name)

    
    year_of_released=store.h3.find("span",class_="lister-item-year text-muted unbold").text.replace('(','').replace(')','')
    year.append(year_of_released)

    
    runtime=store.p.find("span",class_="runtime").text.replace(" min","")
    time.append(runtime)


    rate=store.find("div",class_="inline-block ratings-imdb-rating").text.replace('\n','')
    rating.append(rate)

    
    meta=store.find('span',class_="metascore").text.replace(" ","") if store.find('span',class_="metascore") else "***"
    meta_score.append(meta)

    
    value=store.find_all("span",attrs={"name":"nv"})
    vote=value[0].text
    votes.append(vote)

    
    grosses=value[1].text if len(value)>1 else "###"
    gross.append(grosses)
```

```python
print(movie_name)
print("_______________________________________________________________________________________________________________________")
print(year)
print("_______________________________________________________________________________________________________________________")
print(time)
print("_______________________________________________________________________________________________________________________")
print(rating)
print("_______________________________________________________________________________________________________________________")
print(meta_score)
print("_______________________________________________________________________________________________________________________")
print(votes)
print("_______________________________________________________________________________________________________________________")
print(gross)
```

```python
Movie_DF=pd.DataFrame({"Name of movie":movie_name,"Year of realesed":year,"Duration":time,"Rating":rating,"Metascorce":meta_score,"Vote":votes,"Gross":gross})
```
